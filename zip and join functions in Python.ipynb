# zip() is used to aggregate elements from two or more iterables into a single iterable

```python
mylist1 = [1, 2, 3, 4, 5]
mylist2 = ['a','b','c','d','e']

MyZip = zip(mylist1,mylist2)
print(list(MyZip)) # Convert MyZip to list for printing.
```

```python
# Initializing lists

name = [ "Sha", "Zee", "Mana", "Arsal" ]
roll_no = [ 4, 1, 3, 2 ]
marks = [ 40, 50, 60, 70 ]

# using zip() to map values
mapped = zip(name, roll_no, marks)

# converting values to print as list
mapped = list(mapped)

print(mapped)
```

```python
# unzipping values

names, roll_nos, mark = zip(*mapped)

print(list(names))
print(" ")
print(list(roll_nos))
print(" ")
print(list(mark))

```

# Tuple Unpacking

```python
l=[(2,4),(6,8),(10,12)]

for (t1,t2) in l:
    print(t1)
```

```python
l=[(2,4),(6,8),(10,12)]

for (t1,t2) in l:
    print(t1)
    print(t2)

```

```python
l=[(2,4),(6,8),(10,12)]

for (t1,t2) in l:
    print(t1+t2)
```

# join() is used to join elements of the sequence separated by a string separator.

```python
# Joining with empty separator

list1 = ['h','e','l','l', 'o']

print("".join(list1))
```

```python
# Joining with space separator

list1 = ['h','e','l','l', 'o']

print(" ".join(list1))
```

```python
list1 = ['h','e','l','l', 'o']

print("*".join(list1))
```

```python

myTuple = ("w", "o", "r", "l", "d")

x = "#".join(myTuple)

print(x)
```
