
# lambda handles simple function, def handles larger task

```python
def square(num):
    result=num**2
    return result
```

```python
square(2)
```

```python
square(3)
```

# OR

```python
def square_2(num): return num**2
```

```python
square_2(3)
```

# Creating Function With lambda

```python
square_3 = lambda num: num ** 2   # we use lamda when we need quick results and we dont want to save it for longer time
```

```python
square_3(6)
```

# Grab the first character of the string

```python
first = lambda s: s[0]
```

```python
first('hello')
```

# Adding two number with lambda function

```python
adder = lambda a,b : a + b
```

```python
adder(2,9)
```

```python
adder(-11, -19)
```

# Solving Mathematical Expression (wx + yz)

```python
math_exp = lambda w, x, y, z: w * x   +   y * z
```

```python
math_exp( 2,3,4,5)
```
