### When we have to create our own objects we use class in Python. Class is keyword. Function is called method when put inside the class.
### __init__ is method going to call upon when you create instance of class. self is a keyword which connects
### this method to instance of class. self refer to itself

```python
# Create a class Boy

class Boy(object):                   # we give name (Boy) of the class and input object
    def __init__(self, name):
         self.name = name    # Referencing the attribute (name) of Boy(object)
```

```python
my_boy = Boy(name='Sha')

# In above statement my_boy is an instance of this Boy object type, and it's
# name, that attribute is equal to Sha.
```

```python
my_boy.name  # name is attribute thats why no paranthesis
```

```python
class Boy(object):
    def __init__(self,name, age, grade):
        self.name = name
        self.age = age
        self.grade  = grade

myboy = Boy(name = 'Sha', age = 12, grade = 6)
```

```python
myboy.age
```

```python
myboy.name
```

```python
myboy.grade
```

```python
class Boy01(object):

   # class object attribute. same for any instance of class.

    char = 'strong' # strong is class object attribute. This is not connected to self
                       # but will remain available as class object attribute.

    def __init__(self,name,age,grade):  # class attributes

        self.name = name
        self.age = age
        self.grade  = grade

myboy01 = Boy01(name = 'Sha', age = 12, grade = 6)
```

```python
class Sphere(object):

    pi = 3.14

    def __init__(self,radius = 1):    # Attributes
        self.radius = radius

    def circumference(self):          # Methods
        return (2 * self.pi * self.radius)

    def area(self):
        return 2 * self.pi * (self.radius ** 2)

    def volume(self):
        return 1.33 * self.pi * (self.radius **3)
```

```python
dimension = Sphere()
```

```python
dimension.volume()
```

```python
dimension.area()
```

```python
dimension.circumference()
```

```python
dimension.radius
```

```python
dimension.pi
```
